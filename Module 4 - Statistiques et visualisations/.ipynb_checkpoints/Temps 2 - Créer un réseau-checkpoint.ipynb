### Construire notre réseau

#### Charger des données

```python
import json
with open("./wiki.json","r") as f:
    corpus = json.load(f)
```

```python
G = nx.Graph()
for i in corpus:
    if i in G.nodes:
        G.nodes[i]["weight"]+=1
    else:
        G.add_node(i,weight=1)
    for j in corpus[i]:
        if j in G.nodes:
            G.nodes[j]["weight"]+=1
        else:
            G.add_node(j,weight=1)
        
        if G.has_edge(i,j):
            G[i][j]["weight"]+=1
        else:
            G.add_edge(i,j,weight=1)
```

### Visualiser

https://networkx.org/documentation/stable/reference/generated/networkx.drawing.layout.spring_layout.html#networkx.drawing.layout.spring_layout

```python
pos = nx.spring_layout(G)
```

```python
import matplotlib.pyplot as plt
fix,ax = plt.subplots(figsize=(20,20))
nx.draw(G, pos=pos,node_size=10,with_labels = True)
```

```python
from pyvis.network import Network
net = Network(notebook=True)
net.from_nx(G)
net.show_buttons()
net.show("test.html")
```
