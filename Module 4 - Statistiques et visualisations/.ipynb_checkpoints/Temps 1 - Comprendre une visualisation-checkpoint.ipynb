# Comprendre une visualisation

Partons d'un script complet qui produit une visualisation pour l'analyser

## Visualiser deux variables

Deux objectifs de visualisation : 

1. Un diagramme en barre pour la confiance dans la science
2. Un diagramme pour chacune des variables
3. Un histogramme d'un score de confiance total

### Charger les données

```python
import pandas as pd
import pyshs

data = pd.read_excel("Data-visualisation.xls", index_col=0)
data_a_representer = pyshs.tri_a_plat(data,"La science")["Pourcentage (%)"].drop("Total")
```

### Tracer un diagramme en barre

#### Déclaration implicite avec Pandas

```python
 data_a_representer.plot(kind="bar")
```

#### Déclaration explicite avec Matplotlib

```python
import matplotlib.pyplot as plt

# Déclarer une figure avec un seul élément
fig,ax=plt.subplots(1,1,figsize=(5,3))

# Représenter quelque chose dans la figure (autre manière de faire : utiliser plt.bar)
data_a_representer.plot(kind="bar",ax=ax, rot=20,
                        color=["green","green","orange","orange"], alpha=0.5)

# Paramétrer différents éléments
ax.axes.set_title("Niveau de confiance dans la science")
ax.set_xlabel("Niveau de confiance") # Aussi possible de faire plt.xlabels("Niveau de confiance")
ax.set_ylabel("Pourcentage des répondants") # Aussi possible de faire plt.ylabels("Niveau de confiance")
ax.grid(True)

# Sauvegarder
plt.tight_layout()
plt.savefig("figure1.png",dpi=200)
```

### Tracer une figure avec les différents diagrammes

```python
import matplotlib.pyplot as plt

# Définir une figure avec 10 éléments dedans
fig,ax = plt.subplots(5,2,figsize=(10,10),sharex=True)
ax.shape

# Tracer chaque graphique
for colonne,graph in zip(data.columns,ax.flat):
    data_a_representer = pyshs.tri_a_plat(data,colonne)["Pourcentage (%)"].drop("Total")
    data_a_representer.plot(ax=graph,kind="bar",
                            rot=45,color=["green","yellow","orange","red"],
                            alpha=0.5)
    # Mettre les infos sur chaque graphique
    graph.set_title(colonne)
    graph.set_xlabel("")

# Information générale de la figure
fig.suptitle("Questions concernant la confiance", fontsize=16)
fig.supxlabel("Niveau de confiance") 
fig.supylabel("Pourcentage")    

# Sauvegarder
plt.tight_layout()
plt.savefig("figure2.png",dpi=200)   
```

### Un histogramme du niveau de confiance

Nécessité de construire une nouvelle variable

```python
reco = {'1 - Très confiance':1,'2 - Plutôt confiance':1,
        '3 - Plutôt pas confiance':0,'4 - Pas du tout confiance':0,'Ne sait pas':0}
score = data.replace(reco).sum(axis=1)
```

Afficher

```python
# Construire la figure
fix,ax = plt.subplots(figsize=(10,5))
score.plot(kind="hist",edgecolor='black', linewidth=1.2,ax=ax,title="Histogramme des scores",bins=5)
plt.xlabel("Score")

# Ajouter la moyenne et la médiane
plt.vlines(score.mean(),0,400,color="red")
plt.vlines(score.median(),0,400,color="orange")

# Limite des ordonnées
plt.ylim(0,400)
```
